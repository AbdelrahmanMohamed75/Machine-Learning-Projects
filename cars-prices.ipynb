```python
import pandas as pd 
```

```python
df=pd.read_csv('/kaggle/input/car-price/carprices.csv')
df
```

```python
from sklearn.linear_model import LinearRegression
dummies = pd.get_dummies(df['Car Model']).astype(int)
dummies
```

```python
all_data=pd.concat([df,dummies],axis=1)
all_data
```

```python
all_data.drop(["Car Model"],axis='columns',inplace=True)
```

```python
x=all_data.drop(['Sell Price($)'],axis='columns')
x
```

```python
y=all_data['Sell Price($)']
y
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[25000,4,0,1,0]])
```
