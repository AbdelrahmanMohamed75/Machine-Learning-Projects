```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('/kaggle/input/carprices/carprices.csv')
df
```

```python
dumies=pd.get_dummies(df['Car Model']).astype(int)
dumies
```

```python
merged=pd.concat([df,dumies],axis='columns')
merged
```

```python
 merged.drop(columns='Car Model')
merged
```

```python
x=merged.drop(columns='Sell Price($)')
x
```

```python
y=merged['Sell Price($)']
y
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[45000,4,0,0,1]])
```

```python
model.predict([[86000,7,0,1,0]])
```
